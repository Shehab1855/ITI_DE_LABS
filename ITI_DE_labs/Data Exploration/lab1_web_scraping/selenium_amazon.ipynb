```python
products=x.find_elements("xpath","//h2[@class='a-size-base-plus a-spacing-none a-color-base a-text-normal']/span")
```

```python
all_products = x.find_elements('xpath', "//div[@class='a-section a-spacing-base']")
```

```python
for i in all_products:
    print(i.text)
```

```python

for product in all_products:
    product_text = product.text.split("\n")
    
    # Parse the name, description, price, and rating
    try:
        name = product_text[0]
        price = product_text[-4]

    except IndexError:
        # Handle cases where some data might be missing
        name = product_text[0] if len(product_text) > 0 else "N/A"
        price = product_text[-4] if len(product_text) > 4 else "N/A"
        
    
    # Print the extracted information
    print(f"desc: {name}")
    print(f"Price: {price}")
    print("-" * 40)
```
