```python
from selenium import webdriver
import csv
from selenium.webdriver.common.by import By
import time
```

```python
x=webdriver.Chrome()
```

```python
x.get("https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW")
```

```python
movies_table =x.find_elements("xpath",'//*[@id="table"]/div/table[2]/tbody/tr')
```

```python
for i in movies_table:
    print(i.text)
```

```python
# Extract the movie data into a list of dictionaries
movies_data = []
for movie in movies_table:
    movie_details = movie.text.split("\n")[0].split(" ", 3)
    if len(movie_details) == 4:
        rank, title, lifetime_gross, release_year = movie_details
        movies_data.append({
            "rank": rank.strip(),
            "title": title.strip(),
            "lifetime_gross": lifetime_gross.strip(),
            "release_year": release_year.strip()
        })

# File name
file_name = "movies.csv"

# Write to CSV
with open(file_name, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=["rank", "title", "lifetime_gross", "release_year"])
    writer.writeheader()  # Write the header row
    writer.writerows(movies_data)  # Write the movie data

print(f"Movies saved to {file_name}")
```
