###### Default Palettes:
#### These are basic palettes that come with Seaborn:

"deep": Default palette with 10 distinct, vibrant colors.
"muted": Softer versions of the colors in "deep".
"pastel": Light, pastel shades of the colors in "deep".
"dark": Darker shades of the colors in "deep".
"colorblind": A palette designed to be accessible to people with color vision deficiencies.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("penguins")
data.head()
```

```python
data.info()
```

```python
sns.swarmplot(data=data,x="species",y="body_mass_g",hue='sex',dodge=True ,palette="colorblind",size=5,linewidth=1.5  )
mean = data.groupby("species")["body_mass_g"].mean()
std = data.groupby("species")["body_mass_g"].std()
plt.errorbar(range(len(mean)), mean, yerr=std)
```
