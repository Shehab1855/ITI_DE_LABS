```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
sns.lineplot(data=flights, x='year', y='passengers',errorbar=None)
plt.title('Passengers Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Bar plot for average passengers by month
sns.barplot( data=flights, x='month', y='passengers', estimator='mean', ci=None)
plt.title('Average Number of Passengers per Month')
plt.xlabel('Month')
plt.ylabel('Average Passengers')
plt.show()
```

```python
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap='YlGnBu')
plt.title('Heatmap of Passengers by Year and Month')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
